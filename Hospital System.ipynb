```python
class Patient:
    def __init__(self, name, age, gender, address, contact):
        self.name = name
        self.age = age
        self.gender = gender
        self.address = address
        self.contact = contact
        self.medical_history = []

    def add_medical_record(self, record):
        self.medical_history.append(record)

class MedicalRecord:
    def __init__(self, diagnosis, treatment):
        self.diagnosis = diagnosis
        self.treatment = treatment

class Appointment:
    def __init__(self, patient, doctor, date, time):
        self.patient = patient
        self.doctor = doctor
        self.date = date
        self.time = time

class Doctor:
    def __init__(self, name, speciality):
        self.name = name
        self.speciality = speciality

class Hospital:
    def __init__(self, name):
        self.name = name
        self.doctors = []
        self.patients = []
        self.appointments = []

    def add_doctor(self, doctor):
        self.doctors.append(doctor)

    def add_patient(self, patient):
        self.patients.append(patient)

    def schedule_appointment(self, patient, doctor, date, time):
        # Check if doctor is available
        for appointment in self.appointments:
            if appointment.doctor == doctor and appointment.date == date and appointment.time == time:
                print("Doctor is not available at the specified time.")
                return

        # Check if patient is registered
        if patient not in self.patients:
            print("Patient is not registered. Please add patient first.")
            return

        appointment = Appointment(patient, doctor, date, time)
        self.appointments.append(appointment)
        print(f"Appointment scheduled for {patient.name} with Dr. {doctor.name} on {date} at {time}")

    def generate_invoice(self, appointment):
        # Generate invoice logic
        pass

    def add_medical_record(self, patient, diagnosis, treatment):
        for p in self.patients:
            if p == patient:
                medical_record = MedicalRecord(diagnosis, treatment)
                patient.add_medical_record(medical_record)
                print("Medical record added successfully.")
                return
        print("Patient not found.")

if __name__ == "__main__":
    hospital = Hospital("Super Specialty Hospital")

    # Adding doctors
    doctor1 = Doctor("Dr. John", "Cardiologist")
    doctor2 = Doctor("Dr. Emily", "Neurologist")
    hospital.add_doctor(doctor1)
    hospital.add_doctor(doctor2)

    # Adding patients
    patient1 = Patient("Alice", 35, "Female", "123 Main St", "123-456-7890")
    patient2 = Patient("Bob", 45, "Male", "456 Elm St", "987-654-3210")
    hospital.add_patient(patient1)
    hospital.add_patient(patient2)

    # Adding medical records
    hospital.add_medical_record(patient1, "Hypertension", "Prescribed medication and lifestyle changes")
    hospital.add_medical_record(patient2, "Migraine", "Prescribed pain relievers and rest")

    # Scheduling appointments
    hospital.schedule_appointment(patient1, doctor1, "2024-02-15", "10:00 AM")
    hospital.schedule_appointment(patient2, doctor2, "2024-02-20", "02:00 PM")
```
